```python
#Factorial of a number

n = int(input(" Enter N: "))
fact = 1

while n> 0:
  fact *= n
  n -= 1
print("Factorial =", fact)
```

```python
#Fibonacci Series using while loop

n = int(input(" Enter N: "))
a = 0
b = 1
i = 1
while i<=n:
  print(a)
  c = a + b
  a = b
  b = c
  i += 1
```

```python
limit = int(input("Enter limit: "))
a = 0
b = 1
while a <= limit:
  print(a, end=" ")
  a, b = b, a + b
```

```python
# Reverse a number

num= int(input("Enter a number: "))
rev = 0
while num > 0:
  digit = num % 10
  rev = rev * 10 + digit
  num //=10

print("Reversed number =", rev)
```

```python
#Check palindrome number/string
num = int(input("Enter a number: "))
temp = num
rev = 0
while num > 0:
  rev = rev * 10 + num % 10
  num//= 10

if temp == rev:
  print("Palindrome")
else:
  print("Not palindrome")
```

```python
s = "madam"
i, j = 0, len(s) -  1
is_palindrome = True
while i < j:
  if s[i] != s[j]:
    is_palindrome = False
    break
  i += 1
  j -= 1
if is_palindrome:
  print("Palindrome")
else:
  print("Not palindrome")
```

```python
#check Palindrome number
st =input("enter the string:")
rev = st[::-1]
if st==rev:
  print("palindrome")
else:
  print("not palindrome")
```

```python
#count digits in a number
num = int(input("Enter a number: "))
count = 0
while num > 0:
  count +=1
  num //=10
print("number of digits =", count)
```

```python
my_list = [10, 20, 30, 40, 50]
search_element = int(input("Enter the element to search: "))

i = 0
found = False

while i < len(my_list):
  if my_list[i] == search_element:
    found = True
    break
  i += 1

if found:
  print(f"Element {search_element} found at index {i}.")
else:
  print(f"Element {search_element} not found in the list.")
```
